```python
import pandas as pd
d2012 = pd.read_csv('June 2012 Starbucks_Locations_in_the_US dataset.csv', dtype='str', sep=',')
d2012.head(15)
```

```python
d2012['Zip'] = d2012['Zip'].str[:5]
d2012.head(9)
```

```python
zip2012 = pd.read_csv('count.csv', dtype='str', sep=',')
zip2012.head()
```

```python
df2012_final = pd.merge(d2012, zip2012, left_on = 'Zip', right_on = 'Zip')
df2012_final.head()
```

```python
zip2017 = pd.read_csv('count01.csv', dtype='str', sep=',')
zip2017.head()
```

```python
df2017_final = pd.merge(d2017, zip2017, left_on = 'Postcode', right_on = 'Zip')
df2017_final.head()
```

```python
df2017_final.Postcode.str.split(n=3, expand=True)
```

```python
ZipSplit = df2017_final['Postcode'].str.replace('\s+', '').apply(lambda x: pd.Series(list(x))).add_prefix('Character_')
ZipSplit.head()
```

```python
df2017_final = pd.merge(df2017_final, ZipSplit, left_on = 'Postcode', right_on = '')
df2017_final.head()
```

```python
#d.to_csv("Cleaned_Starbucks_2017.csv")
import csv
df2017_final.to_excel("Cleaned_Starbucks_2017.xls")
```
